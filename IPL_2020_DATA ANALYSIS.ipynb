```python
df['overRuns'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(
    y='overRuns',
    data=df,
    order=df['overRuns'].value_counts().index
)
plt.title("Over Runs")
plt.show()
```

```python
plt.figure(figsize=(10,6))  # Changed from df.figure() to plt.figure()
sns.countplot(
    y='thisOver',
    data=df,
    order=df['thisOver'].value_counts().index
)
plt.title("This Over")
plt.show()
```

```python
df['id'].value_counts()
```

```python
df['thisOver'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.title("This Over")
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```

```python
top: pd.Series = df['speed'].value_counts()
```

```python
plt.figure(figsize=(10,6))
top.plot(kind='bar')
plt.title("Most Over Runs")
plt.xlabel("overRuns")
plt.ylabel("thisOver")
plt.show()
```
